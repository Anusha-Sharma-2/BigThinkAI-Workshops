```python
# numpy functions with given data
data = {
    "Item": ["Banana", "Milk", "Eggs", "Yogurt", "Chips", "Beans", "Tortillas", "Pineapple", "Dip", "Onion", "Garlic", "Tomatoes", "Eggplant", "Fish", "Potatoes", "Okra", "Soy", "Rice" , "Peppers"],
    "Quantity": [1200, 182, 139, 235, 301, 86, 25, 3456, 234, 5678, 123 , 95, 1234, 1234, 6357 , 95, 9687, 967, 43],
    "Cost": [5, 4, 12, 29, 10, 12, 3, 5, 6, 1, 7, 2, 9, 18, 25, 12, 5, 19, 20]
}

df = pd.DataFrame(data)

# # average of COLUMNS
# df.mean(numeric_only=True)
# # df.mean(axis=0)

# # average of ROWS (meaningless in our case)
# # df.mean(axis=1)

# # # sum of COLUMNS
# df.sum(numeric_only=True)

# # # sum specific column
# df["Cost"].sum()

# method to get summary statistics
def get_statistics(df):
    numeric_columns = df.select_dtypes(include=['number'])
    for column in numeric_columns:
        mean = df[column].mean()
        median = df[column].median()
        mode = df[column].mode()
        print(f"{column.capitalize()} Stats:")
        print(f"Mean: {mean}")
        print(f"Median: {median}")
        print(f"Mode: {mode}\n")

get_statistics(df)

# # can also use describe() method
df.describe()
```
